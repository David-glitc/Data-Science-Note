# Linear Regression

## shows relationship between two âœŒ quantities

```python
#linear Regression


import numpy as np 
import matplotlib.pyplot as plt

x = np.random.normal(3,1,1000)
y = 100-(x+np.random.normal(0,0.1,1000))*3
         
plt.scatter(x,y)

def predict(x):
    return -3*x+100
fitline = predict(x)

plt.plot(x,fitline, c='r')
plt.show()
         
```
