```python
%matplotlib inline

from pylab import *

np.random.seed(4)
x= np.random.normal(3.0,1.0,1000)
y = np.random.normal(500.0,10.0,1000)/x

plt.scatter(x,y)
```

```python
import matplotlib.pyplot as plt

a=np.array(x)
b=np.array(y)

p9=np.poly1d(np.polyfit(a,b,9))
xp=np.linspace(0,6,300)
plt.scatter(x,y)
plt.plot(xp,p9(xp),c='r')
plt.show
```
